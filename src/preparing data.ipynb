```python
import pandas as pd
```

```python
df = pd.read_csv('./data/CCLW.csv')
df = df[~df['Document Content URL'].isna()] #240
df
```

```python
df['Category'].value_counts()
```

```python
df[df['Category'] == 'Legislative']['Geographies'].value_counts()
```

```python
df['first_event_year'] = pd.to_datetime(df['First event in timeline'], errors='coerce').dt.year
df['first_event_year']
```

```python
#Check how many instruments

# Split each row into list
split_col = df["Instrument"].fillna("").str.split(r'[|;]')

# Flatten the list of lists
all_elements = [item for sublist in split_col for item in sublist]

from collections import Counter

counts = Counter(all_elements)
counts
```

```python
# Selecting only the legislative documents
legis_df = df[df['Category'] == 'Legislative']
```

```python
split_col = legis_df["Sector"].fillna("").str.split(r'[|;]')

# Flatten the list of lists
all_elements = [item for sublist in split_col for item in sublist]

from collections import Counter

counts = Counter(all_elements)
counts
```

```python
legis_df.to_csv('./data/legis_df.csv', index=False)
```

```python
# Selecting the non-legislative documents
non_legis_df = df[df['Category'] != 'Legislative']
non_legis_df
```

```python
non_legis_df.to_csv('./data/non_legis_df.csv', index=False)
```
